```python
def create_service_context(
  # Constraint parameters
  max_input_size = 4096,        # Context window for the LLM.
  num_outputs = 300,           # Number of output tokens for the LLM.
  chunk_size_limit = 3900,      # Maximum chunk size to use.
  max_chunk_overlap = 30,      # Chunk overlap to use.
  chunk_size = 1024,            # Set chunk overlap to use.
  chunk_overlap_ratio = 0.1    # Chunk overlap as a ratio of chunk size.
  ):

  # Allows the user to explicitly set certain constraint parameters.
  prompt_helper = PromptHelper(max_input_size, num_outputs, chunk_overlap_ratio, chunk_size_limit=chunk_size_limit)

  # LLMPredictor is a wrapper class around LangChain's LLMChain that allows easy integration into LlamaIndex.
  llm_predictor = LLMPredictor(llm=OpenAI(temperature=0, model_name="gpt-3.5-turbo-16k-0613", max_tokens=num_outputs))

  # The embedding model used to generate vector representations of text.
  # define embedding function
  embedding_model = SentenceTransformerEmbeddings(model_name="all-mpnet-base-v2")
  # embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-base-en-v1.5")
  embedding_llm = LangchainEmbedding(langchain_embeddings=embedding_model)

  # Constructs service context
  service_context = ServiceContext.from_defaults(llm_predictor=llm_predictor, prompt_helper=prompt_helper, embed_model=embedding_llm)
  return service_context
```

```python
set_global_service_context(create_service_context())
```

```python
# create client and a new collection
db = chromadb.PersistentClient(path="./chroma_db")
chroma_collection = db.get_or_create_collection("methodology_collection")
```

```python
from tenacity import (
    retry,
    stop_after_attempt,
    wait_random_exponential,
)  # for exponential backoff

@retry(wait=wait_random_exponential(min=1, max=60), stop=stop_after_attempt(6))
def create_llm(**kwargs):
  ### model
  return OpenAI(**kwargs)
```

```python
num_outputs=300
chunk_size=1024
chunk_overlap=30
### Create llm
llm = create_llm(
          temperature=0,
          model_name="gpt-3.5-turbo-16k-0613",
          max_tokens=num_outputs)

text_splitter=TokenTextSplitter(separator=" ",
                                  chunk_size=chunk_size,
                                  chunk_overlap=chunk_overlap)
### metadata extractors
metadata_extractor = MetadataExtractor(
    extractors=[
        TitleExtractor(nodes=5, llm=llm),
        QuestionsAnsweredExtractor(questions=3, llm=llm),
        SummaryExtractor(summaries=["prev", "self"], llm=llm),
        KeywordExtractor(keywords=10, llm=llm),
        ],
    )

# The parser that converts documents into nodes
node_parser = SimpleNodeParser.from_defaults(
      text_splitter=text_splitter,
      metadata_extractor=metadata_extractor)
```

```python
import os
rootdir = './data/'

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        print(os.path.join(subdir, file))
        meth_file_path = os.path.join(subdir, file)
        file_metadata = lambda x : {"filename": x}
        reader = SimpleDirectoryReader(input_files=[meth_file_path], filename_as_id=True)
        documents = reader.load_data()
        print(type(documents))
        # index.insert(document = d, service_context = create_service_context())
        nodes = node_parser.get_nodes_from_documents(documents=documents[3:-2])
        print(f"Doc has node count is {len(nodes)}")
        collection_cnt = chroma_collection.count()
        if collection_cnt > 0:
          print(f"inserting nodes")
          vector_store = ChromaVectorStore(chroma_collection=chroma_collection)
          storage_context = StorageContext.from_defaults(vector_store=vector_store)
          index.insert_nodes(nodes,storage_context=storage_context, service_context=create_service_context())
          print(f"nodes inserted.................. ")
        else:
          print(f"with current Doc it's initial vector store load")
          # set up ChromaVectorStore and load in data
          vector_store = ChromaVectorStore(chroma_collection=chroma_collection)
          storage_context = StorageContext.from_defaults(vector_store=vector_store)
          # When first building the index
          index = VectorStoreIndex(nodes,storage_context=storage_context, service_context=create_service_context())
```
